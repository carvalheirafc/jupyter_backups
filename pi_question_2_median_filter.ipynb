```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
def median_filter(image, kernel_size):
  range_x = len(image)
  range_y = len(image[0])
  radius = kernel_size // 2
  processed_image = np.zeros((range_x, range_y))

  for y in range(range_y-1):
    top = max(y - radius, 0)
    bottom = min(y+radius, range_y-1)
    for x in range(range_x-1):
      left = max(x - radius, 0)
      right = min(x + radius, range_x-1)
      values = []

      for it in range(top, bottom):
        for ik in range(left, right):
          values.append(image[ik][it])
    
      processed_image[x][y] = np.median(values)

  return processed_image
```

```python
im = '2.png'
img = cv2.imread(im)
arr = np.array(img)
plt.imshow(cv2.imread(im))
```

```python
ts = median_filter(arr, 4)
plt.imshow(ts, cmap='gray')
```
